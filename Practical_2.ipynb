```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def step(x):
    return 1 if x >= 0 else 0
```

```python
# OR function
def OR(x1, x2):
    return step(x1 + x2 - 0.5)
```

```python
# NOT function
def NOT(x):
    return step(-x + 0.5)
```

```python
# NOR function
def NOR(x1, x2):
    return step(-x1 - x2 + 0.5)
```

```python
print("=== OR Gate Truth Table ===")
print("Activation: y = step(x1 + x2 - 0.5)")
for x1 in [0,1]:
    for x2 in [0,1]:
        print(f"{x1} {x2} -> {OR(x1,x2)}")
```

```python
print("\n=== NOT Gate Truth Table ===")
print("Activation: y = step(-x + 0.5)")
for x in [0,1]:
    print(f"{x} -> {NOT(x)}")
```

```python
print("\n=== NOR Gate Truth Table ===")
print("Activation: y = step(-x1 - x2 + 0.5)")
for x1 in [0,1]:
    for x2 in [0,1]:
        print(f"{x1} {x2} -> {NOR(x1,x2)}")

print('\n')
```

```python
# OR Decision Boundary
x = np.linspace(-0.25,1.25,200)
y = 0.5 - x
plt.plot(x,y,label="Boundary x1+x2=0.5")
plt.scatter([0,0,1,1],[0,1,0,1],c="red")
plt.title("Decision Boundary for OR Gate")
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()
plt.grid(True)
plt.show()
print('\n')
```

```python
# NOR Decision Boundary
y = -x + 0.5
plt.plot(x,y,label="Boundary -x1-x2+0.5=0")
plt.scatter([0,0,1,1],[0,1,0,1],c="blue")
plt.title("Decision Boundary for NOR Gate")
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()
plt.grid(True)
plt.show()
print('\n')
```

```python
# NOT Step Function
x_vals = np.array([0, 0.5, 0.5, 1])
y_vals = np.array([1, 1, 0, 0])

plt.step(x_vals, y_vals, where="post", linewidth=2)
plt.scatter([0,1], [1,0], c="red", zorder=5, label="Truth values")
plt.title("Activation Function for NOT Gate: step(-x+0.5)")
plt.xlabel("x")
plt.ylabel("Output")
plt.ylim(-0.2, 1.2)
plt.legend()
plt.grid(True)
plt.show()
```
